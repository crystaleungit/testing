```python
html_text = requests.get('https://www.hk01.com/channel/2/%E7%A4%BE%E6%9C%83%E6%96%B0%E8%81%9E').content
hk01_soup = BeautifulSoup(html_text, 'lxml')
# Create empty lists to store the data
titles = []
hrefs = []
timestamps = []

hk01_div = hk01_soup.find_all('div', attrs = {'class':'content-card__main'}) 

# Loop through the data and append to the respective lists
for div in hk01_div:
    a_tag = div.find('a', attrs={'class': 'card-title'})
    b_tag = div.find('div', attrs={'class': 'card-info'})
    if a_tag and b_tag:
        title = a_tag.text.strip()
        href = 'https://www.hk01.com/' + a_tag['href']
        time_text = b_tag.find('time')['datetime']
        
        datetime_obj = datetime.strptime(time_text, '%Y-%m-%dT%H:%M:%S.%fZ')
        timestamp_str = datetime_obj.strftime('%Y-%m-%d %H:%M:%S')
        
        titles.append(title)
        hrefs.append(href)
        timestamps.append(timestamp_str)

# Create a dataframe from the lists
data = {'title': titles, 'href': hrefs, 'timestamp': timestamps}
df = pd.DataFrame(data)
```
